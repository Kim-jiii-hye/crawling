```python
import requests
import re
import urllib.request
from bs4 import BeautifulSoup
import json
```

```python
osulloc_url = 'https://www.osulloc.com/kr/ko/shop/item/list?sort=review'
```

```python
html = urllib.request.urlopen(osulloc_url).read()
soup = BeautifulSoup(html, 'html.parser')
sites = soup.find_all('div', {'class':'item'})
```

```python
link = soup.select_one('#pagination > a:last-child')['href']
totalpage = int(link.split('p=')[1])
print(totalpage)
```

```python
def makeObj(prod, price, imgMain, imgSub):
    return {"prod": prod, "price": price, "imgMain": imgMain, "imgSub": imgSub}
```

```python
data = []

if(totalpage > 1):
    for i in range(0,totalpage+1):
        if(i > 0):
            osulloc_url = 'https://www.osulloc.com/kr/ko/shop/item/list?sort=review&p='+str(i)
            html = urllib.request.urlopen(osulloc_url).read()
            soup = BeautifulSoup(html, 'html.parser')
            sites = soup.find_all('div', {'class':'item'})
        
        for site in sites:
            prod = site.find('p', class_ = 'name').get_text()
            price_bef = site.find('span', class_ = 'price').get_text()
            price = price_bef.replace('\n','')
            imgMain = site.select_one('.thumb > a > .defaultPic > img')['src']
            imgSub = site.select_one('.thumb > a > .hoverPic > img')['src']
            data.append(makeObj(prod, price, imgMain, imgSub))

        json_data = { "data" : data}
        # { “data”: [ { “id”: 1, “item”:””, “imgmain”:””, “imgsub”: “”, “price”: “”, "item_id":""}, ] }

        print(json_data)
```

```python
with open("./listdata.json", "w") as f: 
    json.dump(json_data, f)
```
